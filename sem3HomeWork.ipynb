```python
import pandas as pd
df = pd.read_csv('kc_house_data.csv', sep= ',', encoding= 'windows-1251')
df.head(5)
```

1.1 Создать новый признак price_per_sq_lot,
который будет содержать среднюю стоимость за
один кв. метр общей площади

```python
df['price_per_sq_lot'] = df['price'] / df['sqft_lot']
df['price_per_sq_lot'].head(10)
```

1.2 Создать новый признак delta_renovated,
который будет содержать разницу в годах между
годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

```python
df['delta_renovated'] = (df['yr_renovated'] - df['yr_built']).astype(int).apply(lambda x: x if x > 0 else 0)
df.sample(10)
```

1.3 Создайте признаки года продажи, месяца
продажи


```python
df['yr_saled'] = df['date'].apply(lambda x: x[0:4])
df['month_saled'] = df['date'].apply(lambda x: x[4:6])
df.head()
```

1.4 Удалите признаки date, zipcode, lat, long

```python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
df.sample(10)
```

Составьте несколько сводных таблиц
3.1 Найдите среднюю стоимость домов в
зависимости от количества спален
Отсортируйте от меньшей стоимости к большей


```python
df.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price')
```

3.2 Найдите минимальную, среднюю и
максимальную стоимости домов в зависимости от
состояния дома

```python
df.groupby('condition').agg({'price': ['min', 'mean', 'max']})
```

3.3 Постройте таблицу с подсчетом количества
домов в данных в зависимости от вида на
набережную и оценкой вида

```python
df.pivot_table(index='view',
               columns='waterfront',
               values='price',
               aggfunc='count',
              fill_value=0)
```

3.4 Каких домов в зависимости от этажности и
